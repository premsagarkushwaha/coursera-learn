# 2. Information Retrieval

```python
import numpy as np
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.neighbors import KDTree
```

# imorting dataset

```python
ds= pd.read_csv('log4shell_tweets.csv', usecols=["text", "user_name"])
```

```python
ds   = ds[1:50]
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
tfidf = TfidfVectorizer()
train_tfidf = tfidf.fit_transform(ds.text)
```

```python
ds['tfidf'] = list(train_tfidf.toarray())
```

```python
kdtree = KDTree(train_tfidf.toarray(), leaf_size=3)
```

```python

distance, idx = kdtree.query(ds['tfidf'][ds.user_name =="Brad Wood"].tolist(),k=4)
for i, value in list(enumerate(idx[0])):
    print(f"Name:{ds['user_name'][value]}")
    print(f"Distance: {distance[0][i]}")
    
```
